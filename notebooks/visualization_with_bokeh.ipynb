```python
import numpy as np
import pandas as pd
```

```python
from bokeh.io import output_notebook, push_notebook, output_file, show
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.layouts import column
from bokeh.models.tools import HoverTool
```

```python
index = pd.date_range(start='2019-01-01', periods=100, freq='1d')
df = pd.DataFrame( data={'sin': np.sin(np.linspace(0,20,len(index))), "cos": np.cos(np.linspace(0,20,len(index)))}, index=index)
```

```python
output_notebook()

data_dict = dict(x= pd.to_datetime(df.index))
for df_column in df.columns:
    data_dict[str(df_column)] = df[df_column]

source = ColumnDataSource(data=data_dict)

p_list = []
tools = ["box_select", "hover", "reset", "box_zoom"]
p = figure(x_axis_type="datetime", plot_height=300, plot_width=900, tools=tools)

for df_column in df.columns:

    p = figure(x_axis_type="datetime", plot_height=300, plot_width=900, x_range=p.x_range,  tools=tools, title=str(df_column))
    p.circle("x", str(df_column), source=source)
    p_list.append(p)

show(column(p_list))
```

```python
from sklearn.datasets import load_boston
boston = load_boston()

df = pd.DataFrame(data=boston.data, columns = boston.feature_names)

x = df[df.columns[0]]
y = df[df.columns[1]]
```

```python
output_notebook()

tools = ["box_select", "hover", "reset", "box_zoom"]

p = figure(title="simple line example", plot_height=500, plot_width=500, tools=tools)
r = p.scatter(x, y, color="#8888cc", line_width=1.5, alpha=0.8)

def update(col_x, col_y):
    r.data_source.data['x'] = df[col_x]
    r.data_source.data['y'] = df[col_y]
    push_notebook()
```

```python
from ipywidgets import interact
show(p, notebook_handle=True)
interact(update, col_x=df.columns, col_y=df.columns)
```
